```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
```

## Task II

```python
# 1
uniform = np.random.uniform(0.0, 1.0, 1000)

# 2
norm_1 = np.random.normal(loc=0, scale=1, size=(1000,))

# 3
norm_2 = np.random.normal(loc=0, scale=10, size=(1000,))

# 4
beta_1 = np.random.beta(1, 1, 1000)
beta_2 = np.random.beta(2, 2, 1000)
beta_3 = np.random.beta(3, 9, 1000)

# 5
mean = (0, 0)
cov1 = [[1, 0], [0, 1]]
cov2 = [[2, 1], [1, 2]]
multi_1 = np.random.multivariate_normal(mean, cov1, 10000)
multi_2 = np.random.multivariate_normal(mean, cov2, 10000)
```

## Task III

```python
fig, axs= plt.subplots(2, 3, figsize=(12, 8)) 

# Uniform
axs[0, 0].set_title("Uniform")
x = np.linspace(np.min(uniform), np.max(uniform), 100)
axs[0, 0].plot(x, stats.uniform.pdf(x, *(0,1)), 'r')
axs[0, 0].hist(uniform, density=True)
axs[0, 0].grid()

# Normal I
axs[0, 1].set_title("Normal I")
x_I = np.linspace(np.min(norm_1), np.max(norm_1), 100)
axs[0, 1].plot(x_I, stats.norm.pdf(x_I, *(0, 1)), 'r')
axs[0, 1].hist(norm_1, density=True)
axs[0, 1].grid()

# Normal II
axs[0, 2].set_title("Normal II")
x_2 = np.linspace(np.min(norm_2), np.max(norm_2), 100)
axs[0, 2].plot(x_2, stats.norm.pdf(x_2, *(0, 10)), 'r')
axs[0, 2].hist(norm_2, density=True)
axs[0, 2].grid()

# Beta I
axs[1, 0].set_title("Beta I")
b_1 = np.linspace(np.min(beta_1), np.max(beta_1), 100)
axs[1, 0].plot(b_1, stats.beta.pdf(b_1, *(1, 1)), 'r')
axs[1, 0].hist(beta_1, density=True)
axs[1, 0].grid()

# Beta II
axs[1, 1].set_title("Beta II")
b_2 = np.linspace(np.min(beta_2), np.max(beta_2), 100)
axs[1, 1].plot(b_2, stats.beta.pdf(b_2, *(2, 2)), 'r')
axs[1, 1].hist(beta_2, density=True)
axs[1, 1].grid()

# Beta III
axs[1, 2].set_title("Beta III")
b_3 = np.linspace(np.min(beta_3), np.max(beta_3), 100)
axs[1, 2].plot(b_3, stats.beta.pdf(b_3, *(3, 9)), 'r')
axs[1, 2].hist(beta_3, density=True)
axs[1, 2].grid()
```

## Task IV

```python
v1 = multi_1[:,0]
v2 = multi_1[:,1]
w1 = multi_2[:,0]
w2 = multi_2[:,1]

fig, axs= plt.subplots(1, 4, figsize=(25, 6)) 
axs[0].hist2d(v1, v2, bins = 20, density= True)
axs[1].hist2d(v1, v2, bins = 40, density= True)
axs[2].hist2d(v1, v2, bins = 80, density= True)
sns.histplot(x= v1, y= v2, ax =axs[3])
fig.suptitle("Multivariate I for bins: 20, 40, 80 and seaborn", fontsize="25")

fig, axes= plt.subplots(1, 4, figsize=(25, 6)) 
axes[0].hist2d(w1, w2, bins = 20, density= True)
axes[1].hist2d(w1, w2, bins = 40, density= True)
axes[2].hist2d(w1, w2, bins = 80, density= True)
sns.histplot(x= w1, y= w2, ax =axes[3])
fig.suptitle("Multivariate II for bins: 20, 40, 80 and seaborn", fontsize="25")

plt.show()
```

## Task V and VI

```python
# Task V
alfa = np.random.normal(2, 0.1, 1000)
beta = np.random.normal(0.2, 0.2, 1000)
xn = np.random.normal(0, 3, 1000)
mu = alfa+(beta*xn)
sigma = np.absolute(np.random.normal(0, 1, 1000))
yn = np.random.normal(mu, sigma, 1000)

# Task VI
plt.scatter(xn,yn)
xn = np.linspace(-6, 6)
for i in range(1000):
    yn = alfa[i] + (beta[i]*xn)
    plt.plot(xn, yn, 'orange', alpha = 0.1)
plt.xlim(-6,6)
plt.grid()
plt.show()
```
